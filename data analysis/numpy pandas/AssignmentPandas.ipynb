```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randint(1, 100, size = (6, 4)), columns=['col1', 'col2', 'col3', 'col4'])
print(df)
df.set_index('col1', inplace=True)
print(df)
```

```python
df = pd.DataFrame(np.random.randint(1, 100, size=(3, 3)), columns = ['A', 'B', 'C'], index = ['X', 'Y', 'Z'])
print(df)
element = df.at['Y', 'B']
print(element)
```

```python
df = pd.DataFrame(np.random.randint(1, 100, size = (5, 3)), columns = ['A', 'B', 'C'])
print(df)
df['D'] = df['A'] * df['B']
print(df)
```

```python
df = pd.DataFrame(np.random.randint(1, 100, size= (4, 3)), columns= ['A', 'B', 'C'])
print(df)
row_sum = df.sum(axis=1)
column_sum = df.sum(axis=0)
print(row_sum)
print(column_sum)
```

```python
df = pd.DataFrame(np.random.randint(1, 100, size = (5, 3)), columns=['A', 'B', 'C'])
print(df)
df.iloc[0, 1] = np.nan
df.iloc[2, 2] = np.nan
df.iloc[4, 0] = np.nan
print(df)
df.fillna(df.mean(), inplace=True)
print(df)
```

```python
df = pd.DataFrame(np.random.randint(1, 100, size = (6, 4)), columns=['A', 'B', 'C', 'D'])
print(df)
df.iloc[1, 2] = np.nan
df.iloc[3, 0] = np.nan
df.iloc[5, 1] = np.nan
print(df)
df.dropna(inplace=True)
print(df)
```

```python
df = pd.DataFrame({'Category' : np.random.choice(['A', 'B', 'C'], size = 10), 'Value' : np.random.randint(1, 100, size = 10)})
print(df)
grouped = df.groupby('Category')['Value'].agg(['sum', 'mean'])
print(grouped)
```

```python
df = pd.DataFrame({'Product' : np.random.choice(['A', 'B', 'C'], size = 10), 'Category' : np.random.choice(['A', 'B', 'C'], size = 10), 'Sales' : np.random.randint(1, 100, size = 10)})
print(df)
grouped = df.groupby('Category')['Sales'].sum()
print(grouped)
```

```python
df1 = pd.DataFrame({'Key' : ['A', 'B', 'C', 'D'], 'Value1' : np.random.randint(1, 100, size = 4)})
df2 = pd.DataFrame({'Key' : ['A', 'B', 'C', 'E'], 'Value2' : np.random.randint(1, 100, size = 4)})
print(df1)
print(df2)
merged = pd.merge(df1, df2, on = 'Key')
```

```python
df1 = pd.DataFrame({'A' : np.random.randint(1, 100, size = 3), 'B' : np.random.randint(1, 100, size = 3)})
df2 = pd.DataFrame({'C' : np.random.randint(1, 100, size = 3), 'D' : np.random.randint(1, 100, size = 3)})
print(df1)
print(df2)
row = pd.concat([df1, df2], axis = 0)
print(row)
col = pd.concat([df1, df2], axis = 1)
print(col)
```

```python
date_rng = pd.date_range(start='2022-01-01', end='2022-12-01', freq='D')
df = pd.DataFrame(date_rng, columns=['date'])
df['data'] = np.random.randint(1, 101, size = (len(date_rng)))
df.set_index('date', inplace=True)
print(df)
monthly_mean = df.resample('M').mean()
print(monthly_mean)
```

```python
n = 10;
sum = 0;
count = 1;
while count <= n:
    sum += count;
    count += 1

print("Sum of first", n, "numbers is", sum)
```

```python
count = 0;
for i  in range(2, 10):
    for j in range(2, i):
        if(i % j == 0):
            count += 1
            break
    if(count == 0):
        print(i, end = " ")
    count = 0
```

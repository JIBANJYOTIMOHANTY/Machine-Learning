```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
factorial(5)
```

```python
n = 5;
prod = 1
for i in range(1, n + 1):
    prod *= i;
print("Factorial of", n, "is", prod)
```

```python
sum = 0;
for i in range(1, 11):
    sum += i
print("Sum of first 10 natural numbers is", sum)
```

```python
n = 17
count = 0;
for i in range(1, n + 1):
    if(n % i == 0):
        count += 1;

if(count == 2):
    print(n, "is a prime number")
else:
    print(n, "is not a prime number")
```

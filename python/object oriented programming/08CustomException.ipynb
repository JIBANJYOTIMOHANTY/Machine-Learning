```python
class Error(Exception):
    pass

class dobException(Error):
    pass
```

```python
year = int(input("Enter a year"))
age  = 2024 - year
try:
    if age <= 30 and age >= 20:
        print("Your age is valid")
    else:
        raise dobException
except dobException:
    print("Sorry, Your age is valid")
```

```python
class Animal:
    def speak(self):
        return "Animal"

class Dog(Animal):
    def speak(self):
        return "Dog"
    
class Cat(Animal):
    def speak(self):
        return "Cat"
    
dog = Dog()
cat = Cat()
anml = Animal()
print(dog.speak())
print(cat.speak())

def animal_speak(animal):
    print(animal.speak());
animal_speak(anml)
```

```python
class Shape:
    def area(self):
        return "The area of the figure..............."
    
class Rectangle(Shape):
    def __init__(self, width, height):
        self.height = height;
        self.width  = width;
    
    def area(self):
        return self.height * self.width
    
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
        
    def area(self):
        return 3.14 * self.radius * self.radius
    
def print_area(shape):
    print(f"The area is : {shape.area()}")
    
rectangle = Rectangle(4,5)
circle = Circle(3)
print_area(rectangle)
print_area(circle)
```

```python
from abc import ABC, abstractmethod
class Vehicle(ABC):
    def drive(self):
        print("The vehicles has self drive")
    
    @abstractmethod
    def start_engine(self):
        pass
    
class Car(Vehicle):
    def start_engine(self):
        print("Car engine started.............................")

def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()
    
car = Car()
operate_vehicle(car)
```

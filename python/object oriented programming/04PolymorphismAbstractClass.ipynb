```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def startEngine(self):
        pass
    
class Car(Vehicle):
    def startEngine(self):
        return "Car Engine started......."
    
class Motor(Vehicle):
    def startEngine(self):
        return "Motor Engine started......."
    
    
def print_vehicle(vehicle):
    return vehicle.startEngine()
car = Car()
motor = Motor()
print(print_vehicle(car))
print(print_vehicle(motor))
```

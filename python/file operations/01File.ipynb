```python
with open('example.txt', 'r') as file:
    content = file.read();
    print(content)
```

```python
with open('example.txt', 'r') as file:
    for line in file:
        print(line)
```

```python
with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip())
```

```python
with open('example.txt', 'a') as file:
    file.write("Append operation takes place...", )
```

```python
lines = ["\nfirst line\n", "second line\n", "third line\n"]
with open('example.txt', 'a') as file:
    file.writelines(lines)
```

```python
with open('example.txt', 'r') as file:
    content = file.read()
    
with open('destinationFile.txt', 'w') as file:
    file.write(content)
```

```python
num1 = float(input("Enter first number : "))
num2 = float(input("Enter second number : "))

sum = num1 + num2;
diff = num1 - num2;
prod = num1 * num2;
div = num1 / num2;

print("Sum : ", sum)
print("Difference : ", diff)
print("Product : ", prod)
print("Division : ", div)
print("Modulus : ", num1 % num2)
print("Exponent : ", num1 ** num2)
```

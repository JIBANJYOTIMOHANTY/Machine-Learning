```python
def square(i):
    return i ** 2;
square(5)
```

```python
numbers = [1, 2, 3, 4, 5]
lst = map(square, numbers)
# print(list(map(square, numbers)))
print(list(lst))
```

```python
numbers = [1, 2, 3, 4, 5]
cube = list(map(lambda i : i ** 3, numbers))
print(cube)
```

```python
numbers1 = [1, 2, 3, 4, 5]
numbers2 = [6, 7, 8, 9, 10]
sum = list(map(lambda i, j : i + j, numbers1, numbers2))
print(sum)
```

```python
str = ['1', '2', '3', '4', '5']
numbers = list(map(int, str))
print(numbers)
print(type(numbers[0]))
```

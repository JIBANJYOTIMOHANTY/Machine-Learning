```python
def fib(num):
    if num == 0 or num == 1:
        return num
    return fib(num - 1) + fib(num - 2);
for i in range(11):
    print(fib(i), end=' ')
# print(fib(10))
```

```python
def add_to_dict(a, b = None):
    if b is None:
        b = {}
    b[a] = a ** 2;
    return b;
print(add_to_dict(2))
print(add_to_dict(4, {3:2}))
```

```python
def filter_integers(**args):
    return {i : j for i, j in args.items() if isinstance(j, int)}
print(filter_integers(a = 1, b = 2.0, c = '3', d = 4, e = 5.0, f = 6))
```

```python
def apply(call, lst):
    return [call(i) for i in lst]
print(apply(lambda i : i ** 2, [1, 2, 3, 4, 5]))
print(apply(str, [1, 2, 3, 4, 5]))
```

```python
def outer():
    def inner(x):
        return x ** 2;
    return inner
square = outer()
print(square(5))
print(square(10))
```

```python
def filter_map(filter_func , map_func , lst):
    return [map_func(i) for i in lst if filter_func(i)]

print(filter_map(lambda x : x % 2 == 0, lambda x : x ** 2, [1, 2, 3, 4, 5]));
print(filter_map(lambda x : x > 2, lambda x : x + 1, [1, 2, 3, 4, 5]));
```

```python
def compose(f, g):
    return lambda x : f(g(x));

f = lambda x : x + 1
g = lambda x : x * 2
h = compose(f, g)
print(h(5))  # Output: 11 (5 * 2 + 1)
print(h(10))  # Output: 21 (10 * 2 + 1)
```

```python
from functools import partial
multiply = partial(lambda x, y : x * y, 2)
print(multiply(3))  # Output: 6 (2 * 3)
print(multiply(5))  # Output: 10 (2 * 5)
```

```python
def call(counter = {'count' : 0}):
    counter['count'] += 1
    return counter['count']
print(call())  # Output: 1
print(call())  # Output: 2
print(call())  # Output: 3
```

```python
def max_consecutive_difference(lst):
    # Your code goes here
    pass
    maxDiff = 0;
    for i in range(len(lst) - 1):
        diff = abs(lst[i] - lst[i + 1])
        if diff > maxDiff:
            maxDiff = diff
    
    return maxDiff
print(max_consecutive_difference([1, 2, 3, 4, 5]))  # Output: 1
```

```python
def even_or_odd(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "odd"
```

```python
even_or_odd(-1)
```

```python
def print_numbers(*args):
    for i in args:
        print(i, end = "\t")
print_numbers(1, 2, 3, 4, 5, "Jiban", 3.14, True)
```

```python
def print_details(**args):
    for key, value in args.items():
        print(f"{key} : {value}")
print_details(name = "Jiban", age = 22, city = "Cuttack", country = "India")

```

```python
def printDetails(*args, **kwargs):
    for i in args:
        print(i, end = "\t")
    print()
    for key, value in kwargs.items():
        print(f"{key} : {value}")
printDetails(1, 2, 3, 4, 5, "Jiban", 3.14, True, name = "Jiban", age = 22, city = "Cuttack", country = "India")   
```

```python
def main_welcome(func):
    def sub_welcome():
        print("sub welcome")
        func("Welcome Everyone")
        print("welcome2")
    return sub_welcome()
```

```python
main_welcome(print) 
```

```python
def main_welcome(func, lst):
    def sub_welcome():
        print("sub welcome")
        print(func(lst))
        print("welcome2")
    return sub_welcome()
```

```python
main_welcome(len, [1,2,3,4,5,6,7])
```

```python
def main_welcome(func):
    def sub_welcome():
        print("Welcome1")
        func()
        print("Welcome2")
    return sub_welcome()
```

```python
def course():
    print("Course")
course()
```

```python
main_welcome(course)
```

```python
@main_welcome
def course():
    print("Course")
```

```python
def my_decorator(func):
    def wrapper():
        print("printing 1")
        func()
        print("printing 2")
    return wrapper
```

```python
@my_decorator
def say_Hello():
    print("Hello!")
say_Hello()
    
```

```python
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for i in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_Hello():
    print("Hello")
```

```python
say_Hello()
```

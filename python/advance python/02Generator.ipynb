```python
def square(n):
    for i in range(3):
        yield i ** 2
```

```python
square(3)
```

```python
for i in square(3):
    print(i)
```

```python
a = square(4)
a
```

```python
try:
    next(a)
except StopIteration:
    print("Element ends")
```

```python
def my_generator():
    yield 1;
    yield 2;
    yield 3;
    
```

```python
gen = my_generator()
gen
```

```python
try:
    print(next(gen))
except StopIteration:
    print("Element ends")
```

```python
for i in gen:
    print(i)
```

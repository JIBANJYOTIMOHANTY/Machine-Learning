```python
import sqlite3 as sql
def create_database():
    conn = sql.connect('test.db')
    conn.close()
    print("Database created successfully")
    
create_database()
```

```python
def create_table():
    conn = sql.connect('test.db')
    cursor = conn.cursor()
    # cursor.execute('''drop table employees''')
    cursor.execute('''
                   CREATE TABLE if not exists employees(
                       id Integer PRIMARY KEY,
                       name text not null,
                       age Integer, 
                       department text
                   )
''')
    conn.commit()
    conn.close()
    print("Table created successfully")

create_table()
```

```python
def insert_data(name, age, department):
    conn = sql.connect('test.db')
    cursor = conn.cursor()
    cursor.execute('''
                   INSERT INTO employees(name, age, department)
                   VALUES(?, ?, ?)
''', (name, age, department))
    conn.commit()
    conn.close()
    print("Data inserted successfully")
    
insert_data('Jiban', 22, "Developer")
```

```python
insert_data('Bob', 25, 'Engineering')
insert_data('Charlie', 28, 'Sales')
insert_data('David', 35, 'Marketing')
insert_data('Eve', 22, 'HR')
```

```python
def fetch_data():
    conn = sql.connect('test.db')
    cursor = conn.cursor()
    cursor.execute('''
                   SELECT * FROM employees
''')
    rows = cursor.fetchall()
    for row in rows:
        print(row)
    conn.close()
fetch_data()
```

```python
def ftech_data2(department):
    conn = sql.connect('test.db')
    cursor = conn.cursor()
    cursor.execute('''
                   SELECT * FROM employees WHERE department = ?
''', (department,))
    records = cursor.fetchall()
    for row in records:
        print(row)
    conn.close()
ftech_data2('Engineering')
```

```python
def update_data(id, department):
    conn = sql.connect('test.db')
    cursor = conn.cursor()
    cursor.execute('''
                   UPDATE employees SET department = ? WHERE id = ?
''', (department, id))
    conn.commit()
    conn.close()
    print("Data updated successfully")
update_data(2, 'HR')
fetch_data()
```

```python
# Update the department of 2 employees
update_data(2, 'Research')
update_data(3, 'Customer Support')

# Fetch and display all records
fetch_data()
```

```python
def delete_data(id):
    conn = sql.connect('test.db')
    cursor = conn.cursor()
    cursor.execute('''
                   DELETE FROM employees where id = ?
''', (id, ))
    conn.commit()
    conn.close()
    print("Data deleted successfully")
delete_data(2)
fetch_data()
```

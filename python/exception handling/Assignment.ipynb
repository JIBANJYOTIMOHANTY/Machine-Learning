```python
def div(a, b):
    try:
        result = a / b;
    except ZeroDivisionError as ze:
        print(f"Exception : {ze}")
        result = None;
    finally:
        print("Execution is completed.")
    return result;
print(div(10, 5))
print(div(10, 0))
```

```python
def fun1(lst):
    total = 0
    try:
        total = sum(i + i for i in lst)
    except TypeError as e:
        print(f"Error : {e}")
        total = None
    except Exception as ex:
        print(f"Error : {ex}")
    finally:
        print("Execution was completed.")
    return total
print(fun1([1, 2, 3, 'a']))  # None
print(fun1([1, 2, 3, 4]))  # 10
```

```python
def get_value():
    try:
        value = int(input("Enter a number : "))
    except ValueError as ve:
        print(f"Error : {ve}")
        value = None
    finally:
        print("Execution was completed.")
    return value;
print(get_value())
```

```python
def getDictValue(d, key):
    try:
        value = d[key]
    except KeyError as ke:
        print(f"Error : {ke}")
        value = None
    finally:
        print("Execution was completed.")
    return value;
d = {'a': 1, 'b': 2, 'c': 3}
print(getDictValue(d, 'b'))  # 2
print(getDictValue(d, 'x'))  # None
```

```python
def nested_func1(s):
    try:
        try:
            num = int(s)
        except ValueError as ve:
            print(ve)
            num = None;
        finally:
            print("First execution was completed.")
        if(num is not None):
            try:
                result = 10 / num;
            except ZeroDivisionError as ze:
                print(ze)
                result = None;
            finally:
                print("Second execution was completed.")
            return result
    finally:
        print("All execution was completed.")
print(nested_func1('0'))  # None
print(nested_func1('a'))  # None
print(nested_func1('2'))  # 5.0
```

```python
def indexEle(lst, index):
    try:
        num = lst[index];
    except IndexError as ie:
        print(ie)
        num = None;
    finally:
        print("All execution was completed.")
    return num
lst = [1, 2, 3, 4, 5]
print(indexEle(lst, 2))  # 3
print(indexEle(lst, 10))  # None
```

```python
import requests
def read_url(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.text;
    except requests.RequestException as ex:
        print(f"Network error : {ex}")
        return None;
    finally : 
        print("Execution was completed.")
print(read_url('https://jsonplaceholder.typicode.com/posts/1'))
print(read_url('https://nonexistent.url'))
        
```

```python
import json
def parse_json(json_string):
    try:
        data = json.loads(json_string)
        return data;
    except json.JSONDecodeError as ex:
        print(f"Json error : {ex}")
        return None
    finally:
        print("Execution was completed.")
        
print(parse_json('{"name": "John", "age": 30}'))  # {'name': 'John', 'age': 30}
print(parse_json('Invalid JSON'))  # None
```

```python
class NegativeNumberError(Exception ):
    pass

def check_for_negatives(lst):
    try:
        for num in lst:
            if num < 0:
                raise NegativeNumberError(f"Negative number found : {num}")
    except NegativeNumberError as ne:
        print(f"Error : {ne}")
    finally:
        print("Execution was completed.")
        
check_for_negatives([1, -2, 3, 4])  # Error: Negative number found: -2
check_for_negatives([1, 2, 3, 4])  # Execution complete.
```

```python
def convert_to_integers(lst):
    integers = []
    try:
        for i in lst:
            integers.append(int(i))
    except ValueError as ve:
        print(f"Error : {ve}")
        integers = None
    finally:
        print("Execution was completed.")
    return integers;

print(convert_to_integers(['1', '2', 'three', '4']))  # None
print(convert_to_integers(['1', '2', '3', '4']))  # [1, 2, 3, 4]
```

```python
def convert_with_comprehension(lst):
    try:
        integers = [int(i) for i in lst]
    except ValueError as ve:
        print(f"Error : {ve}")
        integers = None
    finally:
        print("Execution was completed.")
    return integers
print(convert_with_comprehension(['1', '2', 'three', '4']))  # None
print(convert_with_comprehension(['1', '2', '3', '4']))  # [1, 2, 3, 4]
```

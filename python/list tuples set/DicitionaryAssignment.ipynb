```python
nums = {}
for i in range(1, 11):
    nums[i] = i * i
    
print(nums)
```

```python
nums = {i : i * 2 for i in range(1, 11)}
print(nums)
```

```python
cubes = {i : i ** 3 for i in range(1, 11)}
print(cubes)
```

```python
nums = {i : [i * j for j in range(1, 6)] for i in range(1, 6)}
print(nums)
```

```python
d = {i : (i, (i ** 2)) for i in range(1, 6)}
print(d)
```

```python
d = {i : [i, [i ** 2]] for i in range(1, 6)}
print(d)
```

```python
students = {
    "student1" : {
        "name" : "Jiban Jyoti Moahnty", 
        "age"  : 22
    }, 
    "student2" : {
        "name" : "Satabdi Mohanty",
        "age"  : 25
    }
}
print(students)
print(students["student1"]["name"])
print(students["student2"]["name"])
```

```python
students.items()
```

```python
for id, info in students.items():
    print(f"{id} : {info}")
```

```python
for id, student in students.items():
    for key, value in student.items():
        print(f"{key} : {value}")
```

```python
squares = {x : x ** 2 for x in range(1, 11)}
print(squares)
```

```python
even = {x : x ** 2 for x in range(1, 11) if x % 2 == 0}
print(even)
```

```python
numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4];
frequency = {}
for num in  numbers :
    if num in frequency: 
        frequency[num] += 1;
    else:
        frequency[num] = 1;
print(frequency)
```

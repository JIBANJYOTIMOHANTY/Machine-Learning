```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
set3 = set1.union(set2)
print(set3)
set4 = set1.intersection(set2)
print(set4)
set5 = set1.difference(set2)
print(set5)
set6 = set1.symmetric_difference(set2)
print(set6)
print(set1.issubset(set2))
print(set1.issuperset(set2))


```

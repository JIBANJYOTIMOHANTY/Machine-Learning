```python
lst = list(range(1, 21));
print(lst)
```

```python
print(lst[0]);
print(lst[len(lst) // 2])
print(lst[len(lst) - 1])
```

```python
print(lst[:5])
print(lst[5:11])
print(lst[-5:])
```

```python
lst2 = [i ** 2 for i in lst]
print(lst2)
```

```python
even = [i for i in lst if i % 2 == 0]
print(even)
```

```python
import random
nums = [random.randint(1, 100) for i in range(10)]
print(nums)
sort_nums = sorted(nums);
print(sort_nums)
reverse_nums = sorted(nums, reverse = True)
print(reverse_nums) 
unique_nums = sorted(set(nums))
print(unique_nums)
```

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print(matrix)
for i in matrix:
    print(i)
    
for i in matrix:
    for j in i:
        print(j, end = ' ')
    print()
print(matrix[2][2])
```

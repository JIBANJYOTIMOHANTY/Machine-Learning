```python
lst = [78, 82, 94, 95, 96, 87]
average = sum(lst) / len(lst)
print("Average of the list is:", average)
print("Average of the list is: {:.2f}".format(average))
print("Sum of the list is : ", sum(lst))
print("Length of the list is : ", len(lst))
print("Maximum of the list is : ", max(lst))
print("Minimum of the list is : ", min(lst))
print("Sorted list is : ", sorted(lst))
print("Reversed list is : ", sorted(lst, reverse=True))
print("List in ascending order is : ", lst.sort())
print("List in descending order is : ", lst.sort(reverse=True))
```
